```python
df = df.drop_duplicates()
```

```python
df.drop(columns=["Patient ID", "Income", "Continent","Hemisphere"],inplace=True)
```

```python
blood_pressure_dic = {"Systolic": [], "Diastolic": []}
def div_blood(blood_pressure: str):
    split_str = blood_pressure.split("/")
    blood_pressure_dic["Systolic"].append(int(split_str[0]))
    blood_pressure_dic["Diastolic"].append(int(split_str[1]))
```

```python
for i in df["Blood Pressure"]:
    div_blood(i)
```

```python
df["Systolic"] = blood_pressure_dic["Systolic"]
df["Diastolic"] = blood_pressure_dic["Diastolic"]
df.drop(columns=["Blood Pressure"])
```

```python
df["Heart_Attack_Risk"] = df["Heart Attack Risk"]
```

```python
df.drop(columns=["Heart Attack Risk", "Blood Pressure"] , inplace = True)
```

```python
df.columns
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
num_col = ["Heart Rate", "Systolic", "Diastolic"]
```

```python
num_col
```

```python
cat_col = X.select_dtypes(include=object).columns.tolist()
```

```python
cat_col
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
ct = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(), cat_col),
        # ("scaler", StandardScaler(), num_col),
        
    ],
    remainder="passthrough",
)
```

```python
X = ct.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC
svm_classifier = SVC(kernel="rbf", random_state=0)
svm_classifier.fit(X_train, y_train)

y_pred_svm = svm_classifier.predict(X_test)
```

```python
import pickle

with open("heart_attack_xgboost.pkl", "wb") as f:
    pickle.dump(svm_classifier, f)
```
